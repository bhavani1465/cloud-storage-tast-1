```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('sample_data/california_housing_train.csv')
# The original file 'vaccination_data.csv' was not found.
# This line is commented out as the sample data 'california_housing_train.csv' likely does not have a 'date' column.
# df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.head()
```

```python
region_coverage = df.groupby('housing_median_age')[['population', 'median_house_value']].sum().reset_index()
region_coverage = region_coverage.sort_values(by='population', ascending=False)
print(region_coverage)
```

```python
age_vaccination = df.groupby('housing_median_age')[['total_rooms','total_bedrooms']].sum().reset_index()
print(age_vaccination)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 7))
sns.barplot(x='housing_median_age', y='total_rooms', data=age_vaccination, color='skyblue', label='Total Rooms')
sns.barplot(x='housing_median_age', y='total_bedrooms', data=age_vaccination, color='lightcoral', label='Total Bedrooms')
plt.xlabel('Housing Median Age')
plt.ylabel('Count')
plt.title('Total Rooms and Total Bedrooms by Housing Median Age')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
